# Load Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
```

# Load Dataset

```python
sonar = pd.read_csv('sonar data.csv', header=None)
```

```python
sonar.head()
```

```python
sonar.shape
```

```python
sonar.describe()
```

```python
sonar[60].value_counts()
```

M --> MINE  R --> ROCK

```python
sonar.groupby(60).mean()
```

```python
# Separating Data and Labels
X = sonar.drop(columns=60, axis=1)
Y = sonar[60]
```

```python
print(X)
print(Y)
```

# Training and Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Logistic Regression Model

```python
lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

# Model Evaluation

```python
X_train_prediction = lr.predict(X_train)
accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy Score of Model on Training Data is: ', accuracy)
```

```python
X_test_prediction = lr.predict(X_test)
accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy Score of Model on Test Data is: ', accuracy)
```

```python
# Classification Report on Training Data
cr = classification_report(X_train_prediction, Y_train)
```

```python
cr.split('\n')
```

```python
# Classification Report on Testing Data
cr1 = classification_report(X_test_prediction, Y_test)
```

```python
cr1.split('\n')
```

```python
# Confusion Matrix on Training Data
confusion_matrix(X_train_prediction, Y_train)
```

```python
confusion_matrix(X_test_prediction, Y_test)
```
